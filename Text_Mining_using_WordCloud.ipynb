```
import numpy as np
import pandas as pd
# Pillow is a very good Python Image Library
from PIL import Image
# I am importing necessary packages for word cloud
# STOPWORDS is a function that automatically removes words from our results
# Most commonly imported as a seperate file and joined with the data frame
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
%matplotlib inline

# The data set we will be using today is "The Top 500 baby names from 2010 - 2020 in Vienna"
birthnames = pd.read_csv("https://www.wien.gv.at/statistik/ogd/vornamen-top-500.csv", sep=';')
```

```
# Start by selecting column containing the birthnames:
birthnames_column = birthnames['GIVEN_NAME']
# Create a string by joining all elements of 'GIVEN_NAME' with a ',' seperator
birthnames_strings = ','.join(birthnames_column)

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(birthnames_strings)

# Display the generated image:
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```
# Let's take it one step further and say I want to skew the data a little bit I can take Nikola out and replace my name 
birthnames_strings_arno = birthnames_strings.replace('Nikola', 'Arno')

# We are also going to see how the most helpful attributes will make the wordcloud more structured
wordcloud = WordCloud(                      
    background_color='white',
    max_words=50,
    max_font_size=60,
    random_state=0
).generate(str(birthnames_strings_arno))

#Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.margins(x=0, y=0)
plt.show()
```
